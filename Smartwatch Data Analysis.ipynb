```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("D:\Fitabase Data 4.12.16-5.12.16\dailyActivity_merged.csv")
```

```python
data.describe()
df=data
```

```python
plt.xlabel('data.TotalSteps')
plt.ylabel('data.TrackerDistance')
plt.show()
```

```python
fig, ax = plt.subplots()

# Plot TotalSteps and Calories as dots
ax.plot(df['TotalSteps'], range(len(df)), '*', label='Total Steps')
ax.plot(df['Calories'], range(len(df)), '.', label='Calories')

# Set y-axis tick labels to show dates
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['ActivityDate'])

# Add labels and title
ax.set_xlabel('Count')
ax.set_ylabel('Date')
ax.set_title('Fitness Tracker Data')

# Add legend
ax.legend()

# Show the plot
plt.show()
```
